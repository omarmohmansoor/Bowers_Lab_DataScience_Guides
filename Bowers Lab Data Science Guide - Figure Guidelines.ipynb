### **Figure Guidelines**

#### **Nature Journals**

##### Width / Height

Single-column maximum width: 89mm / 3.5 inches
Double-column maximum width: 183mm / 7.2 inches
Height maximum: 170mm / 6.7 (6.69) inches

##### Removing extra white-spaces and saving, usually opt for eps if you need to edit into a facet / subplot in Adobe Illustrator, use PDF if it is ready for publication / single panel

plt.tight_layout()
plt.save_fig(‘name.pdf/eps, dpi=1200 (700 – 1200), bbox_inches=’tight’)

##### Font

All font should be sans-serif, preferably Helvetica or Arial
Panel titles should be at fontsize 8 for multiple-panels, upright, and bold
	plt_set.title(‘..’, fontsize=8, weight=’bold’)
	ax.title(‘..’, fontsize=8, weight=’bold’)
Fontsize for all other texts should be from 5 – 7

#### **Non-Nature Journals**

Follow the same guidelines in terms of height, width, removing white spaces and saving. Only fonts change
Make sure to double check other journal guidelines, JNS / CNS have specific guidelines in terms of sizing of the figures and font sizes. Also, JNS charges per colored figure whether it be gray scale or RGB

##### Fonts

Axis ticks font size = 6-7
Legend font size = 8
Axis labels font size = 8
Title font size = 10-12 (w/ bold)
